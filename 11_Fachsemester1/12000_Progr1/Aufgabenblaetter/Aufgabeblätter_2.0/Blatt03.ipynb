```python
#3.2
import string
def fourletters():
    return_value = []
    consonant_counter = 0
    letters = string.ascii_lowercase
    consonants = 'bcdfghjklmnpqrstvwxyz'
    for letter1 in letters:
        for letter2 in letters:
            for letter3 in letters:
                for letter4 in letters:
                    temp = letter4 + letter3 + letter2 + letter1
                    if (temp[0] in consonants) and (temp[1] in consonants) and (temp[2] in consonants):
                        pass
                    elif (temp[1] in consonants) and (temp[2] in consonants) and (temp[3] in consonants):
                        pass
                    else:
                        return_value.append(temp)
    return return_value


fourletters()
```
